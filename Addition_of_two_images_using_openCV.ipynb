```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
earth = cv2.imread("/content/earth.jpg")
```

```python
earth_rgb = cv2.cvtColor(earth,cv2.COLOR_BGR2RGB)
plt.imshow(earth_rgb);
```

```python
Satellite = cv2.imread("/content/Satellite.png")
```

```python
Satellite_rgb = cv2.cvtColor(Satellite,cv2.COLOR_BGR2RGB)
plt.imshow(Satellite_rgb);
```

```python
print(f"The shape of image 1 and image 2 is {earth.shape} and {Satellite.shape} respectively")
```

```python
earth = np.resize(earth,Satellite.shape)
```

```python
added_images = cv2.add(earth,Satellite)
```

```python
plt.imshow(added_images);
```
